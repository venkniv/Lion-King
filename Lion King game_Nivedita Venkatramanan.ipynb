```python
"""
Created on Wednesday January 26 2022

@author: nivedita.venkatramanan
"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the Disney movie The Lion King (1994). 
    The game consists of three stages and has defined functions for introduction,
    start, the stages, win and fail (which is the user losing the game). 
    In this game, the user is a young lion named Simba who has to cross 
    several stages and win the game to be crowned as King of the Pride Lands.  
    
    No previous knowledge of the movie or its plot is necessary to play the game.
    The game may also help you acquire some knowledge about wildlife.
    
    B) Known Bugs and/or Errors:
    None detected during testing. Any information on bugs is welcome to help modify the code. 
    
    C) The total run-time of the game to win is under 3 minutes. 
    
"""


# Importing some packages for smooth execution of the game

from sys import exit  ## will be used to exit the game 
import random         ## will be used to choose an object at random index from a list
import time           ## Will be used to add a pause at some places to avoid the user from getting too much output at one go

########## Explanation and notes for a few general syntaxes used throughout the game script ::: 
## 1. casefold() -- used to ensure the code reads the input irrespective of case - 
## for a smooth execution of loop with direct match.

## 2. the >> symbol has been at all input prompts to enable the user easily understand that their input comes in that box. 
## It is like giving a direction to the user. 

## 3. The in function has been used in if loops for reading the input responses to ensure that the code correctly
## captures responses even if user types in several words or sentence as reposonse. 

## 4. time.sleep() function - Sleeping the kernel- has been used as a pause or show that the system is 
## thinking for a second at some stage of the game. 

## 5. All the user defined functions start with def - for each stage of the game. 

## 6. All if loops come with an else clause at the end - so as to catch any bugs due to any unexpected inputs.

## 7. \n has been used to go to next line - to make the outputs look appealing to the user by spacing them out well. 

## 8. print('='*100) -- has been used to separate stages and also to make the outputs look appealing to the user 
## by spacing them out well. 
##################################################################################################################

######## Start of the game code ##########

########################################################################################    
####### Defining all the levels of the game, win and fail as the first step ############
########################################################################################

######## Defining UDF for game introduction ##############

def intro(): 
    #### Asking the user if they have watched the movie before. ## saving the input in a variable called movie
    movie = input(prompt = """Have you watched the famous Disney movie called "The Lion King?" \n\n Yes / No >> """).casefold()

    # Start of the if loop       
    if 'yes' in movie or movie == 'y': ## user could choose to just type "y" hence the == condition has been used 
        print("\n\n That's awesome! It will be easy to follow the game! I hope you will like it!! \n\n ")
    
    #else if statement for No
    elif 'no' in movie or movie == 'n': ## user could choose to just type "n" hence the == condition has been used
        print("\n\n Don't worry! I'll help you understand how to go about the game! I hope you like it! \n\n ")

    #else function to catch any bugs. Inputs which are not yes or not will allow the user to try again. 
    else: 
        print('Not a valid response. Respond as Yes or No. Try again.')
        intro() ## goes back to question if the response is neither yes nor no - offering them unlimited chances. 
        
    
    input(prompt = 'Press enter to start! \n')

    print('='*100)     ### Creating a separation to start the game
    
    print(""" \n Short Introduction \n\n 
    
    Mufasa was a huge male lion and the King of the Pride Lands. His envious brother, Scar, had his eyes on the throne. 
    As he loses his chances to become King, after the birth of Mufasa's son, Simba, he plots to kill Mufasa to 
    take over the kingdom. To avoid Simba from becoming the new King after he grows up, he banishes 
    little Simba to a far away land. 
     
    One day, Mufasa's spirit manifests in the clouds and stars and tells him that he had forgotten 
    all his teachings. He tells him that he must go back to his pride, his home.
 
    In this game, you are Simba. Now, after many years, you (Simba) have decided to return to your 
    land. But you need to prove yourself worthy to be crowned King. The goal of this game is that you 
    have to pass all the stages successfully and alive to be able to reach the Pride Rock to become the King.
        
    With this information, lets begin the game, now! 
    
    Good luck! 
    
    """)

    time.sleep(1) ## sleeping the kernel

    input(prompt = "Press enter to continue >> \n")
    
    
    ### Creating a separation to start the game
    print('='*100)     
    
    count=5
    while count > 0:
        print(count)
        time.sleep(0.3)
        count -= 1
    
    print("\n\n GET READY TO PLAY!!! \n\n")
    start_game()  ### Starting the game


######## Defining the UDF for starting the game or Stage 1 - "Face your Fears" #############

def start_game():
    
    print(f"""Stage 1 :: Face your Fears! \n I know you are scared to go back but you have to make a start! \n
    
    As you begin your journey to Pride Lands, leaving your best friends Timon (a meerkat) and 
    Pumba (a wat-hog) behind, you walk for a few miles and come across an elephant graveyard. 
    You see three paths ahead of you and you need to make a decision on which one to take. 
    Each path will be different and will have some surprises. \n""")  ## printing details about the first scene. 
    

    ## Asking for user input 
    print("""  Which path do you wish to take?  

        1) Path 1 
        2) Path 2
        3) Path 3 \n""")
    
    path = input(prompt = " >> ").casefold() ## saving the input as path

    ## defining conditional if loop - for response for path 1 
    if '1' in path or 'Path 1' in path: 
        print("""\n A good start! This is an easy path. There are no obstacles here. You have been lucky today. 
        Now do you want to take rest and sleep or continue the journey? \n""")
        answer = input(prompt = """  1. I want to relax, sleep for sometime. \n 2. I want to continue \n >> """).casefold()
        ## saving the above input in a variable called answer
        
    ## inner conditional loop for path 1 
        if 'relax' in answer or 'sleep' in answer or '1' in answer: 
            print(""""\n You have hardly begun your journey and you want to rest so soon? Don't you want to become a King? """)
            fail()  ## user fails the game for this input
            
      ## elif for another condition for path 1
        elif 'continue' in answer or '2' in answer: 
            print("""\n It is good to know that you are ready to take on this journey. 
            You can continue on the chosen path! Good luck! """)
            survival() ## goes to next stage
            
      ## else for debugging  
        else: 
            print("""Invalid input. You have been logged out of the game.""")
            exit(0) ## exits the game completely
    
    ## elif condition for outer if loop - for path 2 
    elif '2' in path or 'Path 2' in path:
        print("""\n There are a lot of porcupines here. Be careful while you cross the graveyard.""")
        time.sleep(1) ## sleeping the kernel
        print("""\n \n Great! You managed to manoeuvre yourself reasonably well by avoiding the porcupines. 
        You got quilled a few times but it's not that bad. Your saliva contains enzymes that defend infections 
        and also promote a blood clotting mechanism. Lick your wound well and you'll be good to go. \n\n """)
        survival() ## goes to next stage
    
    ## another elif condition for outer if loop - for path 3  
    elif '3' in path or 'Path 3' in path:
        print("""\n Oh no! You have walked into a clan of hyenas!! They have surrounded you now and are ready to 
        attack you. """)
        time.sleep(1) ## sleeping the kernel
        print(""" \n \n You roar at the hyenas in an attempt to defend yourself. You then spot a potential prey and try to 
        move yourself towards it so that you can divert the pack's attention and make your way out safely. \n\n 
        The hyenas notice the prey and leave you to hunt for themselves. You tried to outsmart them and made through this! \n\n """)
        survival() ## goes to next stage
    
    ## else for debugging
    else: 
        print("Invalid input. You have been logged out of the game.")
        exit(0) ## exits the game
        

######## Defining UDF for Stage 2 -- "survival" and "food" ############

def survival(): 
    print('='*100) 
    input(prompt = "Press enter to continue >> ")
    
    ## Setting the stage 2 for the user 
    print("""\n\n Stage 2 :: A struggle for survival 
    
    As you succesfully make your way out of the elephant graveyard, you notice that the Mount Kilimanjaro has erupted! 
    Oh no!! There is a lot of smoke and ash everywhere. Visibility is poor. \n 
    Due to the volcanic eruption, there is a lot of confusion, panic, leading to a stampede. 
    All the animals are hurtling helter skelter. \n\n
    You look around and see that there is little visibility on the LEFT but on the RIGHT you cannot see anything.""") ## introduction for stage2
    
    ## saving the input in a variable called direction
    direction = input(prompt = "\n\n Now what are you going to do? Do you want to take left or right? >> ").casefold() ## taking user input
    
    ### start of if loop for direction
    if 'left' in direction or direction == 'l': ## user could choose to just type "l" hence the == condition has been used
        print("""\n\n Look like a lot of zebras came this side too. Now you are caught in the herd.\n\n  
              Thankfully you have made out alive but your energy levels have dropped tremendously. 
              You will need to re-charge yourself now! \n\n """)
        food() ## goes to food options
    
    ## elif condition for direction
    elif 'right' in direction or direction == 'r': ## user could choose to just type "r" hence the == condition has been used
        print("""\n\n That was a good decision! Due to extremely poor visiblity most animals have avoided this path, so it is easier to walk through this! \n\n""")
        food() ## goes to food options
    
    ## else condition for debugging
    else: 
        print("\n\n Invalid input. You have been logged out of the game.")
        exit(0) ## exits the game
        
        
        
### Defining UDF for food ###

def food(): 
    
    ## Creating a prey list for using the random function #######
    prey_lst = ['camel', 'gazelle', 'wildebeest','zebra']
    prey = random.choice(prey_lst)  ## a random prey will be used as a food option for the below part of the game  

    input(prompt = "Press enter to continue >> ")
    
    print(""" \n\n AH! You have come through the volcanic ash, smoke, stampede! Bravo!!! 
    You must be tired now. You can take some rest and have a good meal so that you can recoup your energy.""")
    time.sleep(2) ## sleeping the kernel
    
    ## user input for food option ## saving the input in a variable called food
    food_option = input(prompt = """ \n\n Oh, there seems to be a carcass of a gazelle lying around here!! 
    Wonder if someone abandoned their hunt or it died of any injury...
    Do you want that (easy food) or would you like to hunt for yourself? \n 1) Easy food \n 2) Hunt \n >> """).casefold()
    
    ### start of if loop for food
    if 'easy' in food_option or 'carcass' in food_option or 'gazelle' in food_option or 'gazel' in food_option or '1' in food_option:
        print("""\n Sorry, this carcass has been poisoned. \n\n It is not easy to become a King. You need to work hard!""")
        fail() ## user fails
    
    ## elif condition for food option
    elif 'hunt' in food_option or '2' in food_option: 
        print("""\n Yes, a lion must hunt for his meal! Let's see what you caught today...""")
        time.sleep(2) ## ## sleeping the kernel - indicating the game is checking what the user caught for a bit -  hence the pause !
        
        ##using the random option for the prey in the output
        print(f"""\n\n Look what you got, you have been successful in hunting a {prey}. 
        Now you can enjoy your meal all by yourself, get some rest before you continue your journey!""")
        war_zone() ### goes to next stage 
    
    ## else for debugging
    else: 
        print("Invalid input. You have been logged out of the game.")
        exit(0) ## exits the game


###### Defining UDF for Stage 3 - War zone #########

def war_zone():
    print('='*100)
    input(prompt = "Press enter to continue >> ") 
    
    print(""" \n\n Stage 3 :: War-Zone
    
    With all that renewed energy, you move ahead and proceed towards Pride Lands. As you are reaching closer to your 
    destination, you see a river flowing. This is the famous Mara river and you have to cross it. \n 
    Even though you are not fond of water, you are an excellent swimmer.""") ## Setting the final stage for the user 
    
    ### Asking the user how they will cross the river ## saving the input in a variable called river
    river = input(prompt = """\n\n What do you want to do? \n 1) Swim \n 2) Take a boat \n >> """).casefold()
    
    ## start of if loop
    if 'swim' in river or '1' in river: 
        print("""\n The crocodiles in this river haven't eaten for a year. This is the migration season and 
        this is what they have been waiting for. You are going to satiate their hunger today... 
        And it's the end of your story.""")
        fail() ## user fails the game
    
    ## elif condition
    elif 'boat' in river or '2' in river: 
        print("""\n Good decision. There are crocodiles in the river, they haven't eaten for a year. 
        This is the migration season and you see them attacking the wildebeest. 
        But they won't attack boats. So you safely cross over to the other side! """)
        win() ## user wins!!!
    
    ## else for debugging
    else: 
        print("Invalid input. You have been logged out of the game.")
        exit(0) ## exits the game


######### UDF for failing the game ##########

def fail():
    print("Sorry, you've failed, but thanks for playing!")  ## printing a fail message for the user. 
    input(prompt = '<Press any key to exit>\n')
    exit(0) ## exits the game 

    
########## UDF for winning the game!!!! ##########    
    
def win(): 
    input(prompt = "Press enter to continue >> ")
    ## printing a winning message for the user
    print("""\n\n\n 
    You have successfully crossed all the stages! Yay! You are now on your way to the Pride Rock in the Pride Lands 
    to re-unite with your family and take over the Kingdom! Congratulations!! \n\n Hakuna Matata!! \n\n """)
    print('='*100)
    print("""
  _    _         _                          
 | |  | |       | |                         
 | |__| |  __ _ | | __ _   _  _ __    __ _  
 |  __  | / _` || |/ /| | | || '_ \  / _` | 
 | |  | || (_| ||   < | |_| || | | || (_| | 
 |_|  |_| \__,_||_|\_\ \__,_||_| |_| \__,_| 
  __  __         _          _               
 |  \/  |       | |        | |              
 | \  / |  __ _ | |_  __ _ | |_  __ _       
 | |\/| | / _` || __|/ _` || __|/ _` |      
 | |  | || (_| || |_| (_| || |_| (_| |      
 |_|  |_| \__,_| \__|\__,_| \__|\__,_|      
                                            
                                            
""")
    print('='*100) 
    input(prompt = '<Press any key to exit>\n')
    exit(0) ## exits the game 

################################################################################################


###### Starting the game ######
#### Asking the user for their name and welcoming them to play the game
## creating a variable "name" that will be used to greet the user. ## saving the input in a variable called name
## User input capitalized with title function as name could be multiple words 
name = input(prompt = 'Hello, what is your first name? >> ').title()  

## printing out welcome message to the user with the variable created 
print(f"\n Hi {name}! Nice to have you here! I have an exciting text adventure game ready for you! \n") 

print('='*100)  ### Design purpose for making the code look neat 
intro() ### first defined function of the game - the game will continue based the functions defined earlier
```
